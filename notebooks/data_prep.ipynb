```python
import pandas as pd
from pathlib import Path
```

```python
file_dir = Path("/Users/admin/AMR/data/raw/data")
output_dir = Path("/Users/admin/AMR/data/cleaned")
```

```python
keep_columns = [
    "topic",
    "geography",
    "stratum",
    "date",
    "metric_value",
    "in_reporting_delay_period"
]

dataframes = []
```

```python
for csv_file in file_dir.glob("*.csv"):
    print("Processing {csv_file.name}")
    df = pd.read_csv(csv_file)
    df = df[keep_columns]
    
    #create organism column from topic
    df["organism"] = (
        df["topic"]
        .str.lower()
        .str.replace("-", "_", regex=False)
        .str.replace(" ", "_", regex=False)
    )

    df = df.rename(columns={"metric_value": "resistance_pct"})

    #remove topic
    df = df.drop(columns=["topic"])

    dataframes.append(df)

amr_total = pd.concat(dataframes, ignore_index=True)

amr_total = amr_total[
        amr_total["in_reporting_delay_period"] == False
    ]

amr_total["date"] = pd.to_datetime(
        amr_total["date"],
        dayfirst = True,
        errors = "coerce"
    )

amr_total = amr_total.sort_values(
        ["organism", "geography", "stratum", "date"]

    )
amr_total = amr_total.drop(
        columns=["in_reporting_delay_period"]
    )
output_path = output_dir / "amr_cleaned.csv"
amr_total.to_csv(output_path, index=False)
print("\n Data preparation complete!")
print("Saved to: ", output_path)
print("\nFinal columns:")
print(amr_total.columns.tolist())
print("\nSample rows:")
display(amr_total.head())
```
